### 1. 2 Bar Plot

```python
data = {'C':20, 'C++':15, 'Java':30, 'Python':35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

plt.bar(courses, values, color ='red', width = 0.4)

plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

# 5. Show an example of Skewed data and removal of skewedness

```python
np.random.seed(42)
skewed_data = np.random.exponential(scale=2, size=1000)

plt.figure(figsize=(8, 5))
plt.hist(skewed_data, bins=30, edgecolor='black')
plt.title('Skewed Data Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
transformed_data = np.log(skewed_data + 1)

plt.figure(figsize=(8, 5))
plt.hist(transformed_data, bins=30, edgecolor='black', color='green')
plt.title('Transformed (Less Skewed) Data Distribution')
plt.xlabel('Transformed Values (Log)')
plt.ylabel('Frequency')
plt.show()
```

# 6. For a sales dataset do a Time Series Visualization

```python
np.random.seed(42)
dates = pd.date_range('2023-01-01', '2023-12-31')
sales = np.random.randint(50, 200, size=len(dates))

sales_data = pd.DataFrame({'Date': dates, 'Sales': sales})
sales_data.set_index('Date', inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(sales_data.index, sales_data['Sales'], label='Sales', color='blue')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

# 7. Build a Scatterplot and suggest dimension reduction

```python
from sklearn.decomposition import PCA

# Sample data with two features (2D)
np.random.seed(42)
data = np.random.rand(100, 2) * 10

# Apply PCA to reduce dimensionality
pca = PCA(n_components=1)
reduced_data = pca.fit_transform(data)

# Original data plot
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1])
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Reduced data plot
plt.subplot(1, 2, 2)
plt.scatter(reduced_data, np.zeros_like(reduced_data))
plt.title('Reduced Dimensionality (PCA)')
plt.xlabel('Principal Component 1')

plt.tight_layout()
plt.show()
```
